```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pyplot
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
fars_df=pd.read_csv("fars_chart.csv", delimiter= "\t") 
fars_df.head()
```

```python
y_total = fars_df['Total']
x_month = fars_df['Month']
farsChart = pyplot.bar(x_month, y_total)
smallMonth_label = ['Jan','Feb','March','April','May','June','July','Aug','Sept', 'Oct', 'Nov','Dec','Total']
pyplot.xticks(ticks = range(len(x_month)), labels=smallMonth_label)
pyplot.tick_params(axis='x', which='major', labelsize=8)
pyplot.tick_params(axis='y', which='major', labelsize=8)
pyplot.title("FARS Chart")
pyplot.xlabel("Months")
pyplot.ylabel("Crashes")
pyplot.show()
```
